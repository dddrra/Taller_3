```python

'''company size y experience level son ordinales. job title son muchas, resumir a 10 o 20, lo mismo con company location,
        salary currency y employee_residence'''
# Creop un diccionario para mapear los valores ordinales de experiencia
experience_mapping = {
    'EN': 1,
    'MI': 2,
    'SE': 3,
    'EX': 4
}

# Aplicar la codificación
df1['experience_level_encoded'] = df1['experience_level'].map(experience_mapping)

#_______________________
company_size_mapping = {
    'S': 1,
    'M': 2,
    'L': 3
}

# Aplicar la codificación
df1['company_size_encoded'] = df1['company_size'].map(company_size_mapping)
```

```python
df1
```

```python
#_______________________________

'''#antes de sacar dummies voy a reducir la cantidad de categorias en  job title son muchas, resumir a 10 o 20, lo mismo con company location,
        salary currency y employee_residence'''
# Seleccionar los 10 job titles más frecuentes
top_job_titles = df1['job_title'].value_counts().head(10).index
df1_no_cat = df1.copy()
#saco company size y experience level
df1_no_cat=df1_no_cat.drop(columns=['company_size','experience_level'])
# Crear una nueva columna donde los job titles no estén en el top 10 se categoricen como 'Otro'
df1_no_cat['job_title_top10'] = df1['job_title'].apply(lambda x: x if x in top_job_titles else 'Otro')
#igual en las otras .--- aca se crea columnas company_location_top10, etc, estas se eliminan con get dummies pero las originales no
top_company_location = df1['company_location'].value_counts().head(10).index
df1_no_cat['company_location_top10'] = df1['company_location'].apply(lambda x: x if x in top_company_location else 'Otro')
top_salary_currency = df1['salary_currency'].value_counts().head(10).index
df1_no_cat['salary_currency_top10'] = df1['salary_currency'].apply(lambda x: x if x in top_salary_currency else 'Otro')
top_employee_residence = df1['employee_residence'].value_counts().head(10).index
df1_no_cat['employee_residence_top10'] = df1['employee_residence'].apply(lambda x: x if x in top_employee_residence else 'Otro')


# Aplicar pd.get_dummies() a la nueva columna
df1_no_cat = pd.get_dummies(df1_no_cat, columns=['job_title_top10'], drop_first=True)
df1_no_cat = pd.get_dummies(df1_no_cat, columns=['company_location_top10'], drop_first=True)
df1_no_cat = pd.get_dummies(df1_no_cat, columns=['salary_currency_top10'], drop_first=True)
df1_no_cat = pd.get_dummies(df1_no_cat, columns=['employee_residence_top10'], drop_first=True)
#sacar las categoricas que ya se dummificaron, company_location etc (estaban desde el inicio, las que borra dummies fueron las columnas top)
df1_no_cat=df1_no_cat.drop(columns=['job_title','company_location','salary_currency','employee_residence'])
# Mostrar las primeras filas del DataFrame resultante
print(df1_no_cat.head())
```
